```python
#### 12. Split 'genres' into multiple columns.
```

```python
#dfmoviegen =  dfmovie["genres"].apply(lambda x: x.split('|'))
dfmoviegen = dfmovie["genres"].str.split("|", n = 5, expand = True) 
dfmovie["genres1"] = dfmoviegen[0]
dfmovie["genres2"] = dfmoviegen[1]
dfmovie["genres3"] = dfmoviegen[2]
dfmovie["genres4"] = dfmoviegen[3]
dfmovie["genres5"] = dfmoviegen[4]
dfmovie.drop(columns =["genres"], inplace = True) 
dfmovie
```

```python
#### 13. Extract year from title e.g. (1995).
```

```python
import re
def get_year(name):
    return re.search('\s\((\d{4})\)$', name).groups()[0]
for name in dfmovie["title"]:
    print (get_year(name))
```

```python
dfmovie["title"].apply(lambda x: get_year(x))
```

```python
#### 14. Select rows based on timestamps later than 2015-02-01.
```

```python
import datetime
def get_date(timestamp):
    datestr = datetime.datetime.fromtimestamp(timestamp)
    return datestr
```

```python
dftag[(dftag["timestamp"].apply(lambda x: get_date(x)) > '2015-02-01')]
 
```
