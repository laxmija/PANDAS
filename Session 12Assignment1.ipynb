2.Show me the average hours per week of all men who are working in private sector

```python
result2=db.execute("select avg([hours-per-week]) from adult where  trim(workclass) ='Private' and trim(sex) = 'Male'")

for row in result2:
    print(row)
```

 3.Show me the frequency table for education, occupation and relationship, separately

```python
# education
myresult = db.execute("SELECT education , COUNT(education) frequency FROM adult GROUP BY education ")

for x in myresult:
    print(x)
```

```python
#   occupation
myresulto = db.execute("SELECT occupation, COUNT(occupation) frequency FROM adult GROUP BY occupation ")

for x in myresulto:
    print(x)
```

```python
#  relationship
myresultr = db.execute("SELECT relationship, COUNT(relationship) frequency FROM adult GROUP BY relationship ")

for x in myresultr:
    print(x)
```

4.Are there any people who are married, working in private sector and having a masters degree

```python
myresultm = db.execute("""SELECT count(*) as "Masters-Married-In-Priv ate" FROM adult where ([marital-status] =" Married-civ-spouse" or [marital-status] =" Married-spouse-absent" or [marital-status] = "  Married-AF-spouse")  and workclass=' Private' and education=' Masters'""") 
res = (int) (myresultm.rowcount) > 0 
print("People those are married working in private sector and having master exis ts is: ", res)
```

5.What is the average, minimum and maximum age group for people working in different sectors

```python
myresulta = db.execute(" SELECT workclass as sector, AVG(age) averageAge, MIN(age) minAge, MAX(age) maxAge  FROM adult GROUP BY workclass ")

for x in myresulta:
    print(x)
```

6.Calculate age distribution by country

```python
cresult = db.execute("select age,[native-country] from adult group by [native-country] ")
for x in cresult:
    print(x)
```
